# Forma de alterar o nome de todas as colunas.
#Alterar o nome de todas colunas utilizando uma espressão regular.
#(Em boas práticas os Nomes de colulas não devem ter letras maiusculas nem barra)

df = df.rename(columns={"ObservationDate":"observacao_data", "Province/State":"provincia_estado", "Country/Region":"pais_regiao", "Last Update":"ultima_atualizacao", "Confirmed":"confirmada", "Deaths":"mortes", "Recovered":"recuperadas"})
df

```python
#Alterar o nome de todas colunas utilizando uma espressão regular.
#(Em boas práticas os Nomes de colulas não devem ter letras maiusculas nem barra)

import re 

def corrige_colunas(col_name):
    return re.sub(r"[/| |]", "", col_name).lower()
```

```python
#TESTE
corrige_colunas("AdgE/P ou")
```

```python

df.columns = [corrige_colunas(col) for col in df.columns]
```
